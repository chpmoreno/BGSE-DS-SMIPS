#### Distribution of the p-value under the null



**1** The c.d.f of continous uniform random variable $x$ is given by:

\begin{equation}
F(x) = \begin{cases} 
      0 & x < a \\
      \frac{x-a}{b-a} & a \leq x \leq b \\
      1 & x > b 
   \end{cases}
\end{equation}


The inverse is then $F^{-1}(y) = a + y (b - a)$ for $0 < y < 1$.

Given $\alpha = 1 - F_H (C_\alpha)$ and $F_H$ is uniform we can use the c.d.f of  continous uniform random variable as follows:

\begin{equation}
\alpha = 1 - \frac{C_\alpha - a}{b - a}
\end{equation}

Then

\begin{equation}
C_\alpha = a + (1 - \alpha)(b - a)
\end{equation}

and taking $y = 1 - \alpha$ I show that

\begin{equation}
C_\alpha = F^{-1}(1 - \alpha) \hspace{2mm} \blacksquare
\end{equation}

**2** Recall Wasserman (2004) Definition 10.11 and Theorem 10.12:
___
*Definition 10.11.* Suppose that for every $\alpha \in (0, 1)$ we have a size $\alpha$ test with rejection region $R_\alpha$. And we have $\Psi$ as a test for single pair hypothesis $\varnothing \neq H \subset \Theta$ (null hypothesis) versus $K = \Theta \setminus H$. Then,

\begin{equation}
p_\Psi(x) = \inf_\left\{ \alpha: T(X) \in R_\alpha \right\} \mathbb{P}^*(T(X) \in R_\alpha)
\end{equation}

That is, the p-value is the smallest level at which we can reject $H$
___
*Theorem 10.12.* Suppose that the size $\alpha$ test is of the form reject $H$ if and only if $T(X) \geq C_\alpha$, then

\begin{equation}
p(x) = \mathbb{P}^*(T(X) \in R_\alpha) = \sup_{\theta \in \Theta} \mathbb{P}_\theta (T(X) \geq T(x))
\end{equation}

where $x$ is the observed value of $X$. If $\Theta = \{\theta_0\}$ then 

\begin{equation}
p(x) = \mathbb{P}_{\theta_0} (T(X) \geq T(x))
\end{equation}
___

Under *Definition 10.11* and *Theorem 10.12*, let the test statistic $T$ fulfill the monotonicity condition

\begin{equation}
\forall \theta_0 \in H: \forall \theta_1 \in K: \forall C_\alpha \in \mathbb{R}: \mathbb{P}_{\theta_0}(T(X) > C_\alpha) \leq \mathbb{P}_{\theta_1} (T(X) > C_\alpha) \hspace{2mm} (1)
\end{equation}

Then call $\Psi$ a test such that

\begin{equation}
\Psi(x) = \begin{cases} 
      1, & T(x) > C_\alpha \\
      0, & T(x) \leq C_\alpha
   \end{cases}(2)
\end{equation} 

and assume that $\mathbb{P}^*$ doesn't depend on $\alpha$. Then it holds

\begin{equation}
p_\Psi(x) = \mathbb{P}^*(T(X) \leq t^*), \hspace{2mm} t^* = T(x) \hspace{2mm} (3)
\end{equation}

To proof the above note that the rejection regions $R_\alpha = (C_\alpha, \infty)$ are nested. Then, $\inf\{\alpha: T(x) \in R_\alpha\}$ is attained in $[t^*, \infty)$. The assertion follows from Definition 10.11 and Theorem 10.12. 

If $H$ is simple, then $\mathbb{P}_H$ continous and $\Psi$ is defined as in (2), and (3) yields then 

$p_\Psi(x) = 1 - F_T(T(x))$, with $F_T$ as in (1). $\blacksquare$

**3** Define $X = F(y)$. Let us define the c.d.f. of $X = F(y)$ as $Fx(x)$. Then, for any $x$ in $[0,1]$ we have:

\begin{align}
Fx(x) &= P[X < x]\\
&= P[F(y) < x]\\
&= P[F^{-1}(F(y)) < F^{-1}(x)]\\
&= P[y < F^{-1}(x)]\\
&= F(F^{-1}(x)) \\
&= x
\end{align}

The distribution that follows the previous c.d.f is the Uniform, hence,

\begin{align*}
F(y) = X \sim Unif
\end{align*}

**4** To prove $p_\Psi (X) \sim_{H} Uniform$ we are going to use **2** and **3**. Therefore we can calculate

\begin{align}
\mathbb{P}_H (p_\Psi (X) \leq t) & = \mathbb{P}_H (1 - F_T(T(X)) \leq t) \\
& = \mathbb{P}_H (F_T(T(X)) \geq 1 - t) \\
& = \mathbb{P} (U \geq 1 - t) \\
& = 1 - \mathbb{P} (U \leq 1 - t) \\
& = 1 - (1 - t) \\
& = t
\end{align}

where $U$ denotes a uniform variate $\blacksquare$.

#### Multiple testing under independence assumptions

**1** $y_1, y_2, ..., y_m$ are $m$ independent random variables, it means:

\begin{equation}
 p(y_1 \cap y_2 \cap ... y_m) = p(y_1) p(y_2) ... p(y_m)
\end{equation}

Due to the independence assumption we have then that other events related to them are also independent. For example we can see that:

\begin{equation}
 p(y_1 > \alpha \cap y_2 > \alpha \cap ... y_m > \alpha] = p(y_1 > \alpha) p(y_2 > \alpha) ... p(y_m > \alpha)
\end{equation}

Then using probability propierties and assuming that $y_m \sim Uniform$, then we have that:

\begin{equation}
 p(y_i > \alpha) = 1 - p(y_i \leq \alpha), \hspace{2mm} i = 1, ..., m
\end{equation}

where $p(y_i \leq \alpha) = \alpha$ as we showed in **Distribution of the p-value under the null [3]**.

Therefore under the independence assumption we have:

\begin{align}
 p(y_1 > \alpha \cap y_2 > \alpha \cap ... y_m > \alpha] =& (1 - \alpha) (1 - \alpha) ... (1 - \alpha) \\
 =& (1 - \alpha)^m
\end{align}

**2** $P(y_i > \alpha) = 1 - \alpha$ is the probability of not reject the null hypothesis. So $\mathbb{P}\left[ \cap_{i = 1}^m \{y_i > \alpha \} \right] = (1 - \alpha)^m$ is the probability of not reject all the null hypothesis. Recall also the exercise before, that is, if there are $m$ independent test and the null hypothesis null are actuallt true, the probability of at least reject one of the test is

\begin{align}
 p(\text{At least one error}) &= 1 - p(\text{no error}) \\
 &= 1 - p(\text{all null hypotesis are not rejected}) \\
 &= 1 - \mathbb{P}\left[ \cap_{i = 1}^m \{y_i > \alpha \} \right] \\
 &= 1 - (1 - \alpha)^m
\end{align}

**3** Let $\alpha$ the overall type I error (reject the null hypothesis when it is true) and let $\alpha_{IND}$ be the individual $\alpha$ of each test then we have the following relationship given the solutions above

\begin{equation}
1 - (1 - \alpha_{IND})^m = \alpha
\end{equation}

So we need to find what is exactly $\alpha_{IND}$:

\begin{align}
1 - (1 - \alpha_{IND})^m =& \alpha \\
1 - \alpha =& (1 - \alpha_{IND})^m \\
(1 - \alpha)^{1/m} =& 1 - \alpha_{IND} \\
\alpha_{IND} =& 1 - (1 - \alpha)^{1/m} 
\end{align}

**4** 

```r
library(ggplot2)
library(reshape2)
library(dplyr)
library(ggthemes)
require(gridExtra)

hypothesis_test_graphs <- function(m){
  alpha    <- seq(0, 1, 0.01)
  f1_alpha <- 1 - (1 - alpha) ^ (1/m)
  f2_alpha <- alpha / m
  f3_alpha <- alpha
  
  data_functions <- data.frame(alpha    = alpha,
                               f1_alpha = f1_alpha,
                               f2_alpha = f2_alpha,
                               f3_alpha = f3_alpha)
  data_functions <- reshape2::melt(data_functions, id.vars = alpha)
  greeks <- list(bquote(f[1](alpha) == 1-(1-alpha)^(1/m)), 
                 bquote(f[2](alpha) == alpha/m), 
                 bquote(f[3](alpha) == alpha))
  
  plot_tot <- ggplot(data = dplyr::filter(data_functions, variable == "f1_alpha"), 
                     aes(x = alpha, y = value, color = variable)) + 
              geom_line(size = 0.8) +
              geom_line(data = dplyr::filter(data_functions, variable == "f2_alpha"), 
                        aes(x = alpha, y = value, color = variable), size = 0.8) +
              geom_line(data = dplyr::filter(data_functions, variable == "f3_alpha"), 
                        aes(x = alpha, y = value, color = variable), size = 0.8) +
              scale_colour_manual(values = c("limegreen", "blue", "red"),
                                  labels = greeks,
                                  guide = guide_legend(title = NULL,
                                                       override.aes = list(linetype = c("solid", "solid", "solid")))) +
              labs(x = bquote(alpha), y = bquote(f(alpha))) +
              ggtitle(bquote(m == .(m))) +
              theme_economist(base_size = 8)
  return(plot_tot)
}
options(repr.plot.width=9, repr.plot.height=10)
grid.arrange(hypothesis_test_graphs(1), hypothesis_test_graphs(5),
             hypothesis_test_graphs(50), hypothesis_test_graphs(100),
             hypothesis_test_graphs(500), hypothesis_test_graphs(1000),
             ncol=2)
```

As we can see in the graphics for $m =$ 1, 5, 50, 100, 500, 1000, $f_2 \leq f_1 \leq f_3$.

**5** We showed in **3** that $\alpha_{IND} = 1 - (1 - \alpha)^{1/m}$. Moreover, we showed also that $1 - (1 - \alpha)^{1/m} \leq \alpha$. It implies that the $\alpha_{IND}$ we use to calculate is always less than the obtained using multiple testing. That means when you select the $\alpha$ for individual tests (usually used) you are selecting it above the p-value necessary to reject the null hypothesis under multiple testing assuming independence, increasing the probability of reject the null hypothesis.
