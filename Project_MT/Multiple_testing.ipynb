#### 1. Distribution of the p-value under the null

**1** The c.d.f of continous uniform random variable $x$ is given by:

\begin{equation}
F(x) = \begin{cases} 
      0 & x < a \\
      \frac{x-a}{b-a} & a \leq x \leq b \\
      1 & x > b 
   \end{cases}
\end{equation}


The inverse is then $F^{-1}(y) = a + y (b - a)$ for $0 < y < 1$.

Given $\alpha = 1 - F_H (C_\alpha)$ and $F_H$ is uniform we can use the c.d.f of  continous uniform random variable as follows:

\begin{equation}
\alpha = 1 - \frac{C_\alpha - a}{b - a}
\end{equation}

Then

\begin{equation}
C_\alpha = a + (1 - \alpha)(b - a)
\end{equation}

and taking $y = 1 - \alpha$ I show that

\begin{equation}
C_\alpha = F^{-1}(1 - \alpha) \hspace{2mm} \blacksquare
\end{equation}

**2** Recall Wasserman (2004) Definition 10.11 and Theorem 10.12:
___
*Definition 10.11.* Suppose that for every $\alpha \in (0, 1)$ we have a size $\alpha$ test with rejection region $R_\alpha$. And we have $\Psi$ as a test for single pair hypothesis $\varnothing \neq H \subset \Theta$ (null hypothesis) versus $K = \Theta \setminus H$. Then,

\begin{equation}
p_\Psi(x) = \inf_\left\{ \alpha: T(X) \in R_\alpha \right\} \mathbb{P}^*(T(X) \in R_\alpha)
\end{equation}

That is, the p-value is the smallest level at which we can reject $H_0$
___
*Theorem 10.12.* Suppose that the size $\alpha$ test is of the form

<center> reject $H$ if and only if $T(X) \geq C_\alpha$,

Then

<center> $p(x) = \mathbb{P}^*(T(X) \in R_\alpha) = \sup_{\theta \in \Theta} \mathbb{P}_\theta (T(X) \geq T(x)) $

where $x$ is the observed value of $X$. If $\Theta = \{\theta_0\}$ then 

<center> $p(x) = \mathbb{P}_{\theta_0} (T(X) \geq T(x)) $
___

Under *Definition 10.11* and *Theorem 10.12*, let the test statistic $T$ fulfill the monotonicity condition

\begin{equation}
\forall \theta_0 \in H: \forall \theta_1 \in K: \forall C_\alpha \in \mathbb{R}: \mathbb{P}_{\theta_0}(T(X) > C_\alpha) \leq \mathbb{P}_{\theta_1} (T(X) > C_\alpha) \hspace{2mm} (1)
\end{equation}

Then call $\Psi$ a test such that

\begin{equation}
\Psi(x) = \begin{cases} 
      1, & T(x) > C_\alpha \\
      0, & T(x) \leq C_\alpha
   \end{cases}(2)
\end{equation} 

and assume that $\mathbb{P}^*$ doesn't depend on $\alpha$. Then it holds

\begin{equation}
p_\Psi(x) = \mathbb{P}^*(T(X) \leq t^*), \hspace{2mm} t^* = T(x) \hspace{2mm} (3)
\end{equation}

To proof the above note that the rejection regions $R_\alpha = (C_\alpha, \infty)$ are nested. Then, $\inf\{\alpha: T(x) \in R_\alpha\}$ is attained in $[t^*, \infty)$. The assertion follows from Definition 10.11 and Theorem 10.12. 

If H is simple, then $\mathbb{P}_H$ continuos and $\Psi$ is defined as in (2) and (3) yields then 

<center> $p_\Psi(x) = 1 - F_T(T(x))$, with $F_T$ as in (1). $\blacksquare$

**3** Define $X = F(y)$. Let us define the c.d.f. of $X = F(y)$ as $Fx(x)$. Then, for any $x$ in $[0,1]$ we have:

\begin{align}
Fx(x) &= P[X < x]\\
&= P[F(y) < x]\\
&= P[F^{-1}(F(y)) < F^{-1}(x)]\\
&= P[y < F^{-1}(x)]\\
&= F(F^{-1}(x)) = x\\
\end{align}

The distribution that follows the previous c.d.f is the Uniform, hence,

\begin{align*}
F(y) = X \sim Unif
\end{align*}

**4** To prove $p_\Psi (X) \sim_{H} Uniform$ we are going to use **2** and **3**. Therefore we can calculate

\begin{align}
\mathbb{P}_H (p_\Psi (X) \leq t) & = \mathbb{P}_H (1 - F_T(T(X)) \leq t) \\
& = \mathbb{P}_H (F_T(T(X)) \geq 1 - t) \\
& = \mathbb{P} (U \geq 1 - t) \\
& = 1 - \mathbb{P} (U \leq 1 - t) \\
& = 1 - (1 - t) \\
& = t
\end{align}

where $U$ denotes a uniform variate $\blacksquare$.
